```python
#merging two data frames
# Creating data frames
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                  index = [1 ,2 , 3, 4])   
```

```python
df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},
                  index = [4, 5, 6, 7])
```

```python
df1 = pd.DataFrame({'key':['K0','K8','K2','K3'],
                   'A':[1,2,3,4],
                   'B':[5,6,7,8]})
df2 = pd.DataFrame({'key':['K0','K1','K2','K3'],
                   'A':[1,12,13,4],
                   'B':[5,16,7,18]})
```

```python
df1.set_index("key",inplace = True)
df2.set_index("key",inplace = True)
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how = "outer",lsuffix = "_x",rsuffix = "_y")
```

```python
df1.reset_index("key",inplace = True)
df2.reset_index("key",inplace = True)
```

```python
df_merge = pd.merge(df1,df2,how = "inner",on = "key")
```

```python
df_merge
```

```python
df_merge = pd.merge(df1,df2,how = "outer",on = "key")
df_merge
```

```python
df_merge = pd.merge(df1,df2,how = "left",on = "key")
df_merge
```

```python
df_merge = pd.merge(df1,df2,how = "right",on = "key")
df_merge
```
